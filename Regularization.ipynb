```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Activation, Dropout
from tensorflow.keras.utils import plot_model
```

```python
# parameters
batch_size = 128
hidden_units = 256
dropout = 0.45
```

```python
model = Sequential()

model.add(InputLayer(input_shape=(1000, ), name='Input_Layer'))
model.add(Dense(256, activation='relu', name='Hidden_Layer_1'))
model.add(Dense(256, activation='relu', name='Hidden_Layer_2'))
model.add(Dense(4, activation='softmax', name='Output_Layer'))

model.summary()
```

```python
plot_model(model, to_file='model.png', show_shapes=True, dpi=100, show_layer_names=True)
```
